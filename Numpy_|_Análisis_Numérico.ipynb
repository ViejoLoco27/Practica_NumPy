```python
import numpy as np
url = '/content/drive/MyDrive/citrus.csv'
np.arange (1,6,1)
data = np.loadtxt(url, delimiter=',', usecols=np.arange (1,6,1), skiprows=1) #Cargar dataset excluyendo la fila 1 y la columna 0
data.shape
```

```python
#Almacenar los datos correspondientes a la col_0 en la variable diameter
#Almacenar los datos correspondientes a la col_1 en la variable weight
diameter = data [0:4999,0]
print (diameter.shape)
weight = data [5000:10000,1]
print (weight.shape)
```

```python
# Oranges

#Almacenar los datos correspondientes a la col_0 en la variable diameter
#Almacenar los datos correspondientes a la col_1 en la variable weight
orange_diameter = data [0:4999,0]
orange_weight = data [0:4999,1]

import matplotlib.pyplot as plt

plt.plot(orange_diameter, orange_weight, 'ro')
plt.xlabel('diameter')
plt.ylabel('weight')
plt.title('Oranges: Diameter v.s. weight')
plt.show()
```

```python
# Grapefruits

#Almacenar los datos correspondientes a la col_0 en la variable diameter
#Almacenar los datos correspondientes a la col_1 en la variable weight
grapefruit_diameter = data [5000:10001,0]
grapefruit_weight = data [5000:10001,1]

import matplotlib.pyplot as plt

plt.scatter(grapefruit_diameter, grapefruit_weight, color='blue', marker='o')
plt.xlabel('diameter')
plt.ylabel('weight')
plt.title('Grapefruit: Diameter v.s. weight')
plt.show()
```

```python
n = np.size(orange_weight)
y = orange_weight
x = orange_diameter
```

```python
a = (n*np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x**2) - np.sum(x)**2)
print (a)

b = (np.sum(y)-a*np.sum(x))/n
print (b)

y = a*x + b
print (y)
```

```python
plt.plot(orange_diameter, orange_weight, 'ro')
plt.plot(x, y, 'b-')
plt.xlabel('diameter')
plt.ylabel('weight')
plt.title('Oranges: Diameter v.s. weight')
plt.show()
```

```python
n_1 = np.size(grapefruit_weight)
y_1 = grapefruit_weight
x_1 = grapefruit_diameter
```

```python
a_1 = (n_1*np.sum(x_1*y_1)-np.sum(x_1)*np.sum(y_1))/(n*np.sum(x_1**2) - np.sum(x_1)**2)
print (a_1)

b_1 = (np.sum(y_1)-a_1*np.sum(x_1))/n_1
print (b_1)

y_1 = a_1*x_1 + b_1
print (y_1)
```

```python
plt.plot(grapefruit_diameter, grapefruit_weight, 'ro')
plt.plot(x_1, y_1, 'b-')
plt.xlabel('diameter')
plt.ylabel('weight')
plt.title('Grapefruits: Diameter v.s. weight')
plt.show()
```
